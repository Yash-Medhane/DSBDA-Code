```python
import pandas as pd

# Load dataset
df = pd.read_csv("data/salaries.csv")
df
```

```python
# a. Create data subsets
subset1 = df[['company', 'job']]
subset2 = df[['degree', 'salary_more_then_100k']]
subset1, subset2
```

```python
# b. Merge Data (horizontal merge for demo)
merged_data = pd.concat([subset1, subset2], axis=1)
merged_data
```

```python
# c. Sort Data by company name
sorted_data = df.sort_values(by='company')
sorted_data
```

```python
# d. Transposing Data (first 3 rows only)
transposed = df.head(3).transpose()
transposed
```

```python
# e. Shape and reshape Data (pivot example)
original_shape = df.shape
reshaped = df.pivot_table(index='company', columns='salary_more_then_100k', values='job', aggfunc='first')
original_shape, reshaped
```
