```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("data/Toyota.csv")
df.head(5)
```

```python

# Clean basic issues (e.g., remove rows with missing Price, Age, KM, or FuelType for plotting)
df = df.dropna(subset=['Price', 'Age', 'KM', 'FuelType'])
```

```python
# a. Age vs Price - Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Price', hue='FuelType', palette='deep')
plt.title('Scatter Plot: Age vs Price')
plt.xlabel('Age (Years)')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
# b. Distribution of KM data - Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['KM'], bins=30, kde=True, color='red')
plt.title('Histogram: Distribution of KM')
plt.xlabel('Kilometers Driven')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# c. FuelType wise Cars Count - Bar Plot
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='FuelType', palette='Set2')
plt.title('Bar Plot: FuelType-wise Car Count')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.grid(True)
plt.show()
```
