```python
import pandas as pd

# Read the dataset
df = pd.read_csv("data/Toyota.csv")
df.head(5)
```

```python
# a. Create Subset selecting columns 'Price', 'Age', 'FuelType', 'Automatic'
subset_a = df[['Price', 'Age', 'FuelType', 'Automatic']]
subset_a
```

```python
# b. Create subset having records of all Petrol vehicles
subset_b = df[df['FuelType'] == 'Petrol']
subset_b
```

```python
# c. Create subset of cars’ data having Price greater than 15000 and Age less than 8 years
subset_c = df[(df['Price'] > 15000) & (df['Age'] < 8)]
subset_c
```

```python
# d. Merge records of the above two data subsets
merged_subset = pd.concat([subset_b, subset_c]).drop_duplicates()
merged_subset
```
