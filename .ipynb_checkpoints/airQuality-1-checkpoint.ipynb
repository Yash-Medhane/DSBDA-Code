```python
import pandas as pd

# a. Read data and display summary
df = pd.read_csv("data/airquality.csv")
df.head(5)
```

```python
df_summary = df.describe(include='all')
df_summary  # 🔹 Summary of the dataset
```

```python
# b. Create data subsets
subset1 = df.iloc[11:50]  # Observations from index 11 to 49
subset2 = df[df['PM2.5'] < 60]  # Temperature not present, using PM2.5 < 60
subset1  # 🔹 Subset 1
subset2  # 🔹 Subset 2
```

```python
# c. Merge subsets
merged = pd.concat([subset1, subset2])
merged  # 🔹 Merged Data
```

```python
# d. Sort data by PM2.5 (as closest to Temp in air quality context)
sorted_df = df.sort_values(by='PM2.5')
sorted_df  # 🔹 Sorted Data
```
