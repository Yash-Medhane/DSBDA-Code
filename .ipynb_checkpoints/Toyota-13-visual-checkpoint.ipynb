```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dataset
df = pd.read_csv("data/Toyota.csv")
df.head(5)
```

```python

# a. Scatter plot - Car Price by Age
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Price', data=df)
plt.title('Car Price by Age')
plt.xlabel('Age (Years)')
plt.ylabel('Price')
plt.show()
```

```python
# b. Histogram on Cars data KM
plt.figure(figsize=(8, 6))
sns.histplot(df['KM'], kde=True, bins=10)
plt.title('Histogram of KM (Kilometers)')
plt.xlabel('Kilometers')
plt.ylabel('Frequency')
plt.show()
```

```python
# c. Bar plot on counts of FuelType category (Petrol, Diesel, and CNG)
plt.figure(figsize=(8, 6))
sns.countplot(x='FuelType', data=df)
plt.title('Counts of Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.show()
```
