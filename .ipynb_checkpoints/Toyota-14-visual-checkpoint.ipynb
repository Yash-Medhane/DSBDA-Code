```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("data/Toyota.csv")
df.head(5)
```

```python
# a. Age vs Price - Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Price'], color='blue', alpha=0.5)
plt.title('Age vs Price Scatter Plot')
plt.xlabel('Age (Years)')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
# b. Distribution of KM data - Histogram
plt.figure(figsize=(8, 6))
plt.hist(df['KM'], bins=30, color='green', edgecolor='black', alpha=0.7)
plt.title('Distribution of KM (Kilometers) Driven')
plt.xlabel('Kilometers')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# c. FuelType wise CarsCount - Bar plot
plt.figure(figsize=(8, 6))
df['FuelType'].value_counts().plot(kind='bar', color='purple', edgecolor='black', alpha=0.7)
plt.title('Fuel Type Wise Cars Count')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```
