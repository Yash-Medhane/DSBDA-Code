```python
import pandas as pd

# Read the dataset
df = pd.read_csv("data/Toyota.csv")
df.head(5)
```

```python
# a. Sort observations on Price values order
df_sorted = df.sort_values(by='Price', ascending=True)
df_sorted  # Display the sorted data
```

```python
# b. Create Subset by Selecting columns 'Price', 'Age', 'FuelType', 'Automatic' and first 10 records
subset_b = df.loc[:9, ['Price', 'Age', 'FuelType', 'Automatic']]
subset_b  # Display the subset data
```

```python
# c. Create subset of cars data having Price greater than 15000 and Age less than 8
subset_c = df[(df['Price'] > 15000) & (df['Age'] < 8)]
subset_c  # Display the subset data
```

```python
# d. Create subset of cars data consuming Petrol
subset_d = df[df['FuelType'] == 'Petrol']
subset_d  # Display the subset data
```

```python
# e. Apply decimal normalization on Price column
df['Price_normalized'] = df['Price'].apply(lambda x: x / 10000)
df[['Price', 'Price_normalized']]  # Display the original and normalized Price columns
```
