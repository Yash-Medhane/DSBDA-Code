```python
import pandas as pd

df = pd.read_csv('data/housing.csv')
df.head(5)
```

```python
# a. Subset where income > 5 and avg rooms < 6
a = df[(df['Avg. Area Income'] > 5) & (df['Avg. Area Number of Rooms'] < 6)]
a
```

```python
# b. Merge with regional lookup (mock example)
region = pd.DataFrame({
    'Address': df['Address'].head(),
    'Region': ['North', 'South', 'East', 'West', 'Central']
})
b = pd.merge(df, region, on='Address', how='left')
b
```

```python
# c. Sort by Price and Area Population
c = df.sort_values(by=['Price', 'Area Population'])
c
```

```python
# d. Transpose summary statistics
d = df.describe().transpose()
d
```

```python
# e. Pivot table: avg Price across income and house age bins
df['IncomeBin'] = pd.cut(df['Avg. Area Income'], bins=3)
df['AgeBin'] = pd.cut(df['Avg. Area House Age'], bins=3)
e = df.pivot_table(index='IncomeBin', columns='AgeBin', values='Price', aggfunc='mean')
e
```
